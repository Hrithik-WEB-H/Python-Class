```python
import pandas as pd
```

```python
data = [11,12, 13, 14]
series = pd.Series(data)
print(series)
```

```python
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)
print(df)
```

```python
#### Creating Datagram

data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)
print(df)
```

```python
#### Accessing data in dataframe

print(df['Name'])
print(df.iloc[1])
print(df.loc[1, "Name"])
```

```python
#### Data manipulation

#### Adding and removing data

df = pd.DataFrame({"A": [1,2,3], "B": [25, 30, 35]})
df["C"] = [7,8,9]
print(df)
df = df.drop('B', axis = 1)
print(df)
```

```python
#### Aggregation\

df = pd.DataFrame({'A': [1,2,3], 'B': [4,5,6]})
print(df.aggregate('sum'))
```

```python
df = pd.DataFrame({'category':['cse', 'cse', 'eee', 'eee'], 'Salary': [25, 30, 35, 40]})
print(df)
grouped = df.groupby("category")
print(grouped)
```

```python
##### Aggregation data
df = pd.DataFrame({'A': [1,2,3], 'B': [25, 30, 35]})
print(df.aggregate('sum'))
print(df.aggregate('min'))
```
